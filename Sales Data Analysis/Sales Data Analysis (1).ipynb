# The Following Bar graph Shows that the TotalSales by Stores;

```python
plt.figure(figsize=(10,5))
plt.bar(Sales['Store'],Sales['TotalSales'])
plt.title(" Store vs TotalSales ")
plt.xlabel("Store")
plt.ylabel("TotalSales")
plt.show()
```

# import Seaborn library;

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

# The following relplot Shows that The Total Sales by Stores;

```python
plt.figure(figsize=(16,18))
sns.relplot(x='Store',y='TotalSales',kind='line',data=Sales)
plt.show()
```

# The following Stripplot Shows that The TotalSales by Stores;

```python
sns.set_style("darkgrid")
sns.stripplot(x="Store",y="TotalSales",jitter=0.50,dodge=True,data=Sales)
plt.show()
```

# The following catplot Shows that The TotalSales by Stores;

```python
sns.catplot(x='Store',y='TotalSales',kind='boxen',data=Sales)
plt.show()
```

# The following violinplot Shows that The TotalSales by Stores;

```python
sns.violinplot(x='Store',y='TotalSales',data=Sales)
plt.show()
```

# The following swarmplot Shows that The TotalSales by Stores;

```python
sns.swarmplot(x='Store',y='TotalSales',dodge=True,data=Sales)
plt.show()
```

# The following boxplot Shows that The TotalSales by Stores;

```python
sns.boxplot(x='Store',y='TotalSales',width=0.7,data=Sales)
plt.show()
```

# The following catplot Shows that The TotalSales by Stores;

```python
sns.catplot(x='Store',y='TotalSales',kind='swarm',data=Sales)
plt.show()
```
